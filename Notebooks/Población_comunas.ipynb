```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import warnings

warnings.filterwarnings('ignore')


```

# Realizamos Web Scraping

```python
# Define la URL de la página de Wikipedia que contiene la información sobre las comunas de Buenos Aires.
Web_scrap = 'https://es.wikipedia.org/wiki/Anexo:Comunas_de_la_ciudad_de_Buenos_Aires_por_poblaci%C3%B3n'

# Utilizar la biblioteca requests para obtener el contenido de la página web definida por la URL.
respuesta = requests.get(Web_scrap)

# Utilizar BeautifulSoup para parsear el contenido HTML obtenido de la página web, usando 'html.parser' como el parser.
soup = BeautifulSoup(respuesta.text, 'html.parser')

# Utilizar pandas para leer todas las tablas HTML encontradas en la página web y almacenarlas.
pag = pd.read_html(Web_scrap)

# Imprimr el primer DataFrame de la lista, que corresponde a la tabla que queremos utilizar.
pag[0]
```

```python
# Asignar el primer DataFrame de la lista a una nueva variable o DataFrame
Poblacion_CABA = pag[0]
```

```python
# Imprimir todos los nombres de columnas en Poblacion_CABA
print(Poblacion_CABA.columns.tolist())
```

```python
# "Aplanamos" el DataFrame
Poblacion_CABA.columns = ['_'.join(col).strip() for col in Poblacion_CABA.columns.values]
```

```python
Poblacion_CABA
```

```python
# Cambiar los nombres de las columnas
nuevos_nombres = [
    "Puesto (2022)", 
    "Comuna", 
    "CENSO 1991", 
    "CENSO 2001", 
    "CENSO 2010", 
    "CENSO 2022", 
    "Crecimiento 2010-2022", 
    "Unnamed: 7_level_1"
]

# Asignar los nuevos nombres a las columnas del DataFrame
Poblacion_CABA.columns = nuevos_nombres
```

```python
Poblacion_CABA
```

```python
# Eliminar la fila con índice 15
Poblacion_CABA.drop(index=15, inplace=True)

```

```python
# Eliminar las columnas que no vamos a utilizar
Poblacion_CABA.drop(columns=['Puesto (2022)', 'Crecimiento 2010-2022', 'Unnamed: 7_level_1'], inplace=True)


```

```python
# Limpiar las columnas y convertirlas a integer
for columna in ['CENSO 1991', 'CENSO 2001', 'CENSO 2010', 'CENSO 2022']:
    # Eliminar espacios no separables y espacios regulares
    Poblacion_CABA[columna] = Poblacion_CABA[columna].str.replace('\xa0', '').replace(' ', '')
    # Convertir a integer
    Poblacion_CABA[columna] = Poblacion_CABA[columna].astype(int)

# Verificar los cambios en los tipos de datos
print(Poblacion_CABA.dtypes)
```

```python
Poblacion_CABA
```

```python
# Guardar el DataFrame modificado en un archivo .csv
Poblacion_CABA.to_csv('../DataSets/Poblacion_CABA.csv', index=False)

print("DataFrame modificado y guardado como 'DataSets/Poblacion_CABA.csv'")
```
