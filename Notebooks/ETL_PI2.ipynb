## Primera Parte: Extracción de datos e importación de librerías y módulos

```python
# Importar librerías y módulos necesarios
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
# Ruta al archivo Excel
archivo_excel = '../DataSets/homicidios.xlsx'

# Leer la hoja "HECHOS" del archivo Excel
df_hechos = pd.read_excel(archivo_excel, sheet_name='HECHOS', engine='openpyxl')

# Leer la hoja "VICTIMAS" del archivo Excel
df_victimas = pd.read_excel(archivo_excel, sheet_name='VICTIMAS', engine='openpyxl')
```

## Tercera Parte: Fusión de DataFrames

```python
# Realizar un merge (fusión) entre df_hechos y df_victimas
df_merged = pd.merge(df_hechos, df_victimas, left_on='ID', right_on='ID_hecho', how='inner')

# Mostrar las primeras filas del DataFrame resultante para verificar
print(df_merged.head())

# Mostrar información del DataFrame resultante para revisar tipos de datos y no nulos
print(df_merged.info())
```

## Cuarta Parte: Limpieza de Datos

Este paso incluye la eliminación de columnas redundantes o no necesarias y el renombramiento de columnas para claridad y consistencia, preparando el DataFrame para análisis posteriores.


Al revisar la información precedente, se puede observar que al realizar el "merge", las columnas 'ID_hecho', 'FECHA_y', 'AAAA_y', 'MM_y', 'DD_y' y 'VICTIMA_y' se encuentran repetidas.

La información de la columna "FECHA_FALLECIMIENTO" ya se encuentra en las varibales "DD", "MM" y "AAAA".

La columna "Altura" cuenta con demasiados nulos, por lo que no se puuede rescatar su información imputando algún valor.

La información de la columna "PARTICIPANTES" ya se encuentra en las columnas "VICTIMA" y "ACUSADO".

El contenido de la columna "XY (CABA)" ya se encuentra en las variables "pos x" y "pos y".

Finalmente, los valores de las columnas 'Dirección Normalizada', 'Cruce' y 'Calle' se encuentran en la columna "LUGAR_DEL_HECHO".





```python
# Eliminar columnas redundantes
df_merged_cleaned = df_merged.drop(columns=['ID_hecho', 'FECHA_y', 'AAAA_y', 'MM_y', 'DD_y', 'VICTIMA_y', 'FECHA_FALLECIMIENTO', 'Altura', 'PARTICIPANTES','XY (CABA)', 'Dirección Normalizada', 'Cruce', 'Calle', 'FECHA_x'])

# Renombrar las columnas restantes para evitar confusión, eliminando el sufijo _x en las columnas necesarias
df_merged_cleaned.rename(columns={
    'AAAA_x': 'AAAA',
    'MM_x': 'MM',
    'DD_x': 'DD',
    'VICTIMA_x': 'VICTIMA'
}, inplace=True)

```

```python
df_merged_cleaned
```

## Quinta Parte: Normalización y Conversión de Datos

```python
# Reemplazamos los valores " '', 'None', 'SD' y None" con la expresión "nan"
df_merged_cleaned.replace(to_replace=['', 'None', 'SD', None], value=np.nan, inplace=True)
```

```python
# Normalizar el texto a minúsculas para las columnas categóricas seleccionadas
columns_to_normalize = ['VICTIMA', 'ACUSADO', 'ROL', 'SEXO', 'TIPO_DE_CALLE']
df_merged_cleaned[columns_to_normalize] = df_merged_cleaned[columns_to_normalize].apply(lambda x: x.str.lower())

# Revisar los valores únicos después de la normalización para identificar inconsistencias
for column in columns_to_normalize:
    print(f"Valores únicos en '{column}':", df_merged_cleaned[column].unique())
```

```python
# Convertir la columna "COMUNA" a tipo object
df_merged_cleaned['COMUNA'] = df_merged_cleaned['COMUNA'].astype('object')
```

```python
# Convertir la columna "EDAD" de float a integer 
df_merged_cleaned['EDAD'] = pd.to_numeric(df_merged_cleaned['EDAD'], errors='coerce').astype('Int64')
```

```python
# Eliminar la columna "HORA"
df_merged_cleaned.drop(columns=['HORA'], inplace=True)

# Renombrar "HH" a "HORA"
df_merged_cleaned.rename(columns={'HH': 'HORA'}, inplace=True)

# utilizaremos 'Int64' (tipo entero nullable de pandas) que soporta NaN.
df_merged_cleaned['HORA'] = pd.to_numeric(df_merged_cleaned['HORA'], errors='coerce').astype('Int64')
```

## Sexta Parte: Tratamiento de Duplicados

Podemos apreciar que no hay duplicados.

```python
# Identificar duplicados
duplicados = df_merged_cleaned.duplicated()

# Mostrar los duplicados
print("Filas duplicadas:")
print(df_merged_cleaned[duplicados])
```
